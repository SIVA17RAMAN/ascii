```python
import numpy as np
x=np.array([4,8,13,7])
y=np.array([11,4,5,14])
xm=np.mean(x)
ym=np.mean(y)
print(xm,ym)
covxy=np.cov(x,y)
print(covxy)
w,v=np.linalg.eig(covxy)
print(w)
print(v)
vt=v.transpose()
print(vt)
e1,e2=np.hsplit(vt,2)
print(e1)
print(e2)
x=x-xm
y=y-ym
data=np.stack((x.T,y.T),axis=0)
print(data)
p1=e1*data
print(p1)
p2=e2*data
print(p2)
import matplotlib.pyplot as np
np.scatter(p1,p2)
np.show()
```
